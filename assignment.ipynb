# Here we find the desired table from the wikipedia page

```python
from bs4 import BeautifulSoup
import requests

headers=[]
cells = []
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source,'lxml')
i = 0
for columns in soup.findAll('th')[:3]:    ##here we try to find tables headers
    if len(headers) <= 3:
        headers.append(columns.text)
headers[2] = headers[2].replace("\n","")
for columns in soup.findAll('tr'):       ##here we try to find tables cells contents
    cells.append([])
    for values in columns.findAll('td'):
        cells[i].append(values.text)
    if len(cells[i]) != 3:               ##remove unwanted columns from other tables since our table contains 3 cells per row
        cells.remove(cells[i])
        i+=0
    else:
        cells[i][2] = cells[i][2].replace("\n","")
        i+=1
            
```

# create the dataframe from the exported data, change the name of the first column

```python
import pandas as pd


table = pd.DataFrame(cells,columns=headers)
table = table.rename(columns = {"Postcode" : "PostalCode"})
```

```python
table
```

# drop the rows containing the Not assigned values

```python
table.drop(table.loc[table['Borough']=='Not assigned'].index, inplace=True)
```

# reset index

```python
table = table.reset_index(drop = True)
```

```python
table
```

# We create a list of lists, each list contains a postal code, the borough and the neighborhoods seperated by commas

```python
grouped = []
codes = list(table.groupby('PostalCode').groups.keys())
Bo_neigh = list(table.groupby('PostalCode').groups.values())
for i in range(len(codes)):
    grouped.append([])
    grouped[i].append(codes[i])
    grouped[i].append(table.iloc[Bo_neigh[i][0],1])
    grouped[i].append([])
    for j in range(len(Bo_neigh[i])):
        grouped[i][2].append(table.iloc[Bo_neigh[i][j],2])
    neigh = str(grouped[i][2])
    char_to_rv = ['[',']','\'']
    for elt in char_to_rv:
        neigh = neigh.replace(elt,'')
    grouped[i][2] = neigh
```

# New dataframe as we want

```python
table_clean = pd.DataFrame(grouped,columns=headers)
```
